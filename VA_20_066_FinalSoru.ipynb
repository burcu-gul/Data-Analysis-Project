```python
import numpy as np

# Aykırı değerleri belirlemek için IQR yöntemi kullanıldı
def detect_outliers_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

numerical_df = df.select_dtypes(include=[np.number])

# Aykırı değerleri tespit edildi (id numarası bulunan nitelik hariç)
outliers = detect_outliers_iqr(numerical_df.drop(columns=["1000025"]))

# Aykırı değerler ilgili sütunların ortalamasıyla dolduruldu
for col in numerical_df.columns:  
    if col != "1000025":  
        col_mean = numerical_df[col][~outliers[col]].mean()  # Aykırı olmayan değerlerin ortalaması
        df.loc[outliers[col], col] = col_mean

print(df.head())
```

```python
from sklearn.preprocessing import MinMaxScaler
#Özellik seçimi için filtreleme yöntemlerinden düşük varyans filtresi kullanıldı.Bu yüzden veri seti öncelikle normalize edildi.
# Veri setine min-max normalizasyonu uygulandı 
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(df)

# Normalleştirilmiş veri seti DataFrame'e dönüştürüldü
normalized_df = pd.DataFrame(normalized_df, columns=df.columns)

print(normalized_df.head())
```

```python
from sklearn.feature_selection import VarianceThreshold

# Düşük varyans filtresi için eşik değeri belirleme (örneğin, 0.01)
threshold = 0.01

# Düşük varyansa sahip özellikleri filtrelemek için VarianceThreshold kullanıldı
selector = VarianceThreshold(threshold=threshold)
selected_features = selector.fit_transform(normalized_df)

# Filtrelenmiş özelliklerin indekslerini alarak, orijinal veri setinde ilgili sütunlar seçildi
selected_feature_indexes = selector.get_support(indices=True)
selected_features_df = normalized_df.iloc[:, selected_feature_indexes]

# Filtrelenmiş özellikleri kontrol etme
print(selected_features_df.head())
```

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

# Tüm niteliklerin normal QQ plotlarını çizildi
for column in df.columns:
    if df[column].dtype != 'object':  # Sayısal nitelikleri kontrol ediliyor
        stats.probplot(df[column], dist="norm", plot=plt)
        plt.title(f"QQ Plot for {column}")
        plt.xlabel("Theoretical Quantiles")
        plt.ylabel("Ordered Values")
        plt.show()
```

```python
#niteliklerin normal dağılıma sahip olmadığı için spearsman korealsyon yöntemi kullanıldı
#Bu çıktı, her değişken çifti için Spearman korelasyon katsayılarını içerir. Pozitif değerler, değişkenler arasında pozitif bir ilişki olduğunu gösterirken, negatif değerler negatif bir ilişkiyi ifade eder. Değer ne kadar yakınsa, ilişki de o kadar güçlüdür.
#Örneğin, 1.3 ve 2.1 değişkenleri arasındaki korelasyon katsayısı oldukça yüksektir (0.765704), bu da bu iki değişken arasında güçlü bir pozitif ilişki olduğunu gösterir.
#Ancak, 1.5 değişkeni, diğer değişkenlerle korelasyon analizi yapılamaz çünkü tüm değerleri eksik görünüyor. Bu durumda bu değişken diğerlerini etkilemiyor olabilir veya bu değişkenin veri setiniz için önemsiz olduğunu gösterebilir.

#Sonuçları değerlendirirken, korelasyon katsayılarının anlamlılığını ve işaretlerini dikkate almak önemlidir.
spearman_corr = df.corr(method='spearman')
print(spearman_corr)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from scipy.stats import f_oneway


# Sınıf etiketi ayrı bir değişkende saklandı
y = df["2.1"]

# İlk ve son sütunları hariç özellikleri ayırma
X = df.iloc[:, 1:-1]

# Veri setini normalize etme (id sütunu hariç)
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Veri kümesini eğitim ve test kümelerine bölmek (normalize edilmiş veri ile)
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Algoritma seçimi ve eğitimi
models = {
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Naive Bayes": GaussianNB()
}

# Hiperparametre ayarlaması
param_grids = {
    "KNN": {'n_neighbors': [3, 5, 7, 9]},
    "Random Forest": {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20]},
    "Naive Bayes": {}
}

for name, model in models.items():
    param_grid = param_grids[name]
    grid_search = GridSearchCV(model, param_grid, cv=5)
    grid_search.fit(X_train, y_train)
    print(f"{name}: En iyi parametreler: {grid_search.best_params_}")

# Model değerlendirmesi
accuracies = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    accuracies.append(acc)
    print(f"{name} accuracy: {acc}")

# ANOVA testi
f_statistic, p_value = f_oneway(*[model.predict(X_test) for name, model in models.items()])
print(f"ANOVA F-Statistic: {f_statistic}, p-value: {p_value}")

# En iyi modelin değerlendirilmesi
best_model_name = list(models.keys())[accuracies.index(max(accuracies))]
best_model = models[best_model_name]

best_model.fit(X_train, y_train)
y_train_pred = best_model.predict(X_train)
y_test_pred = best_model.predict(X_test)

train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
f1 = f1_score(y_test, y_test_pred, pos_label=2.0)  # 2.0 pozitif sınıf etiketi olarak kullanıldı, değişebilir
precision = precision_score(y_test, y_test_pred, pos_label=2.0)
recall = recall_score(y_test, y_test_pred, pos_label=2.0)

print(f"En iyi model ({best_model_name}) için:")
print(f"Eğitim seti accuracy: {train_accuracy}")
print(f"Test seti accuracy: {test_accuracy}")
print(f"F1 score: {f1}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")

# Final model oluşturma 
final_model = best_model
print(final_model)
```

```python

# mcnemar test
from statsmodels.stats.contingency_tables import mcnemar
# define contingency table
table = [[129, 54], 
         [9, 13]]
# calculate mcnemar test
result = mcnemar(table, exact=True)
# summarize the finding
print('statistic=%.3f, p-value=%.3f' % (result.statistic, result.pvalue))
alpha = 0.05
if result.pvalue > alpha:
	print('Same proportions of errors (fail to reject H0)')
else:
	print('Different proportions of errors (reject H0)')

# Example of calculating the mcnemar test
from statsmodels.stats.contingency_tables import mcnemar
```

Bu kod, McNemar testini gerçekleştirmek için kullanılmıştır ve testin sonuçlarını yorumlamak için adım adım yapılandırılmıştır. İlk olarak, statsmodels kütüphanesinin contingency_tables modülünden mcnemar fonksiyonu içe aktarılır. Bu fonksiyon, McNemar testini gerçekleştirir. Daha sonra, test için kullanılan kontenjans tablosu tanımlanır. Bu tablo, iki sınıflandırıcı modelin hata oranlarını karşılaştırır. Tabloda yer alan 129 ve 13 değerleri, her iki modelin de doğru sınıflandırdığı örneklerin sayısını temsil ederken, 54 değeri ilk modelin yanlış, ikinci modelin doğru sınıflandırdığı örneklerin sayısını, 9 değeri ise ilk modelin doğru, ikinci modelin yanlış sınıflandırdığı örneklerin sayısını temsil eder.

Kodda, mcnemar fonksiyonu kullanılarak McNemar testi gerçekleştirilir ve exact=True parametresi ile küçük veri setleri için kesin bir p-değer hesaplaması sağlanır. Testin sonucunda elde edilen istatistiksel değer (statistic) ve p-değeri (p-value) yazdırılır. Anlamlılık düzeyi (alpha) 0.05 olarak belirlenmiştir. P-değeri bu eşik değerden büyükse, H0 hipotezi reddedilmez ve hata oranlarının aynı olduğu kabul edilir. Ancak, p-değeri bu eşik değerden küçükse, H0 hipotezi reddedilir ve hata oranlarının farklı olduğu kabul edilir.

Çıktıya göre, statistic=9.000 ve p-value=0.000 olarak hesaplanmıştır. P-değeri 0.05'ten küçük olduğu için (0.000 < 0.05), H0 hipotezi reddedilir ve iki modelin hata oranlarının istatistiksel olarak anlamlı bir şekilde farklı olduğu sonucuna varılır. Bu sonuç, iki modelin performanslarının farklı olduğunu gösterir. Bu, sınıflandırıcı modellerin hatalarının aynı oranda olmadığını ve bir modelin diğerinden daha iyi veya daha kötü performans gösterdiğini belirtir.

McNemar testi, özellikle birbiriyle ilişkili iki sınıflandırma algoritmasının performansını karşılaştırmak için kullanılır. Örneğin, birbirine alternatif iki algoritma varsa ve bunların performanslarını karşılaştırmak istiyorsak, McNemar testi kullanılabilir.Burada da knn ve random forest algoritmaları kullanıldığı için mcnemar testi tercih edilmiştir

Ancak, eğer kıyaslanan algoritmaların sayısı daha fazlaysa ve sıralı veya sırasız kategorik verilerle çalışılıyorsa, Freidman testi daha uygun olabilir. Freidman testi, çoklu algoritma karşılaştırmalarında ve bu algoritmaların performanslarını sıralamak için kullanılır.

Sonuç olarak McNemar testi, yalnızca iki algoritmanın performansını karşılaştırmak için önerilirken, Freidman testi, üç veya daha fazla algoritmanın performansını karşılaştırmak için daha uygun bir seçenektir. Bu nedenle, eğer iki algoritmanın performansını değerlendirmek istiyorsak, McNemar testi daha uygun bir seçenek olabilir.
